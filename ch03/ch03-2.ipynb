```python
img = cv2.imread('/content/gdrive/My Drive/colab/images/girl.jpg')
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
gimg = cv2.cvtColor(RGB_img, cv2.COLOR_BGR2GRAY)
print(gimg.shape)
plt.imshow(gimg, cmap='gray')
```

```python
gPlus = np.zeros((gimg.shape[0], gimg.shape[1]), dtype=np.ubyte)
```

```python
val = 100

for h in range(gimg.shape[0]):
  for w in range(gimg.shape[1]):
    plusVal = gimg[h, w] + val
    if(plusVal > 225):
      plusVal = 225
    gPlus[h, w] = plusVal
plt.imshow(gPlus, cmap='gray')
```

```python
img = cv2.imread('/content/gdrive/My Drive/colab/images/aero2.bmp')
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
gimg1 = cv2.cvtColor(RGB_img, cv2.COLOR_BGR2GRAY)
print(gimg1.shape)
plt.imshow(gimg1, cmap='gray')
```

```python
img = cv2.imread('/content/gdrive/My Drive/colab/images/lena256.jpg')
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
gimg2 = cv2.cvtColor(RGB_img, cv2.COLOR_BGR2GRAY)
print(gimg2.shape)
plt.imshow(gimg2, cmap='gray')
```

```python
for h in range(gimg1.shape[0]):
  for w in range(gimg1.shape[1]):
    plusVal = np.int32(gimg1[h, w]) + gimg2[h, w]
    if(plusVal > 225):
      plusVal = 225
    gPlus[h, w] = plusVal

plt.imshow(gPlus, cmap='gray')
```

```python
for h in range(gimg1.shape[0]):
  for w in range(gimg1.shape[1]):
    plusVal = (np.int32(gimg1[h, w]) + np.int32(gimg2[h, w]))/2
    if(plusVal > 225):
      plusVal = 225
    gPlus[h, w] = plusVal

plt.imshow(gPlus, cmap='gray')
```

```python
weight = 0.7

for h in range(gimg1.shape[0]):
  for w in range(gimg1.shape[1]):
    plusVal = np.int32(gimg1[h, w])*weight + np.int32(gimg2[h, w])*(1-weight)
    if(plusVal > 225):
      plusVal = 225
    gPlus[h, w] = plusVal

plt.imshow(gPlus, cmap='gray')
```
